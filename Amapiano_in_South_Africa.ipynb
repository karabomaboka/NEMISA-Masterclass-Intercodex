```python
import pandas as pd
import numpy as np

# Define the number of records you want in your dataset
num_records = 50

# Create a list of artist names
artists = ["DJ Maphorisa", "Kabza De Small", "Sha Sha", "MFR Souls", "De Mthuda", "Kamo Mphela", "Vigro Deep", "Kwesta", "Focalistic"]

# Create a list of Amapiano song titles
song_titles = ["Amantombazane", "eMcimibini", "Love You Tonight", "Amapiano Anthem", "Umlilo", "Ke Star", "Kokota", "Vula Vala", "eWallet"]

# Generate random release dates within a range (e.g., last 2 years)
start_date = "2021-01-01"
end_date = "2022-12-31"
release_dates = pd.to_datetime(np.random.choice(pd.date_range(start_date, end_date), num_records))

# Generate random popularity scores (e.g., on a scale of 1 to 100)
popularity_scores = np.random.randint(1, 101, num_records)

# Create a DataFrame to store the Amapiano music data
amapiano_df = pd.DataFrame({
    "Artist": np.random.choice(artists, num_records),
    "Song Title": np.random.choice(song_titles, num_records),
    "Release Date": release_dates,
    "Popularity Score": popularity_scores
})

# Display the first few records of the dataset
print(amapiano_df.head())
```

```python
# Check for missing data
missing_data = amapiano_df.isnull().sum()
print("Missing Data:\n", missing_data)

# Remove duplicates
amapiano_df = amapiano_df.drop_duplicates()

# Convert 'Release Date' to datetime
amapiano_df['Release Date'] = pd.to_datetime(amapiano_df['Release Date'])
```

```python
# Summary statistics
summary_stats = amapiano_df.describe()
print("Summary Statistics:\n", summary_stats)
```

```python
# Group data by artist and calculate the mean popularity score
artist_popularity = amapiano_df.groupby("Artist")["Popularity Score"].mean().sort_values(ascending=False)
print("Artist Popularity:\n", artist_popularity)
```

```python
import matplotlib.pyplot as plt

# Group data by release year and count the number of songs released each year
release_date_trends = amapiano_df.groupby(amapiano_df['Release Date'].dt.year)['Song Title'].count()

# Create a line plot to visualize release trends
plt.figure(figsize=(10, 6))
plt.plot(release_date_trends.index, release_date_trends.values, marker='o')
plt.title("Amapiano Music Releases Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Songs Released")
plt.grid(True)
plt.show()
```

```python
# Sort the dataset by popularity score and display the top songs
top_songs = amapiano_df.sort_values(by="Popularity Score", ascending=False).head(10)
print("Top Amapiano Songs:\n", top_songs)
```

```python
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Create a bar chart for artist popularity
plt.figure(figsize=(12, 6))
sns.barplot(x=artist_popularity.values, y=artist_popularity.index, palette="viridis")
plt.title("Popularity of Amapiano Artists")
plt.xlabel("Average Popularity Score")
plt.ylabel("Artist")
plt.show()
```

```python
# Create a line plot to visualize release trends
plt.figure(figsize=(10, 6))
plt.plot(release_date_trends.index, release_date_trends.values, marker='o')
plt.title("Amapiano Music Releases Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Songs Released")
plt.grid(True)
plt.show()
```

```python
# Create a bar chart for top songs
plt.figure(figsize=(12, 6))
sns.barplot(x="Popularity Score", y="Song Title", data=top_songs, palette="magma")
plt.title("Top Amapiano Songs by Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Song Title")
plt.show()
```

```python
# Fictional data for a geographical heatmap
locations = ["Johannesburg", "Cape Town", "Durban", "Pretoria", "Port Elizabeth", "Bloemfontein"]
popularity = [75, 45, 60, 80, 30, 20]

# Create a fictional geographical heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(np.array(popularity).reshape(1, -1), cmap="coolwarm", annot=True, xticklabels=locations, yticklabels=False, cbar=False)
plt.title("Amapiano Music Popularity Across South African Cities (Fictional)")
plt.show()
```

```python
# Group data by release year and calculate the mean popularity score for songs released in each year
song_popularity_trends = amapiano_df.groupby(amapiano_df['Release Date'].dt.year)['Popularity Score'].mean()

# Create a line plot to visualize song popularity trends
plt.figure(figsize=(10, 6))
plt.plot(song_popularity_trends.index, song_popularity_trends.values, marker='o', color='orange')
plt.title("Amapiano Song Popularity Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Average Popularity Score")
plt.grid(True)
plt.show()
```

```python
# Calculate the count of songs per artist
song_count_per_artist = amapiano_df['Artist'].value_counts()

# Create a scatter plot of artist popularity vs. song count
plt.figure(figsize=(10, 6))
sns.scatterplot(x=song_count_per_artist, y=artist_popularity, hue=artist_popularity, palette="coolwarm", s=100)
plt.title("Artist Popularity vs. Song Count")
plt.xlabel("Number of Songs Released")
plt.ylabel("Average Popularity Score")
plt.legend(title="Popularity")
plt.grid(True)
plt.show()
```
